```python
#Import libraries
import shap
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('c:/users/hlmas/Desktop/MSc/Ethics/MSc-Ethics/wine_data.csv')
df.head()
```

```python
df["high_quality"] = 0
df.loc[df["quality"]>5,"high_quality"] = 1
```

```python
df.sample(10)
```

```python
df['high_quality'].value_counts()
```

```python
# Separate Features and Target Variables
X = df.drop(columns=['quality','high_quality'])
y = df['high_quality']

# Create Train & Test Data
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,stratify =y,random_state = 13)

# Build the model
rf_clf = RandomForestClassifier(max_features=2, n_estimators =100 ,bootstrap = True)

rf_clf.fit(X_train, y_train)

# Make prediction on the testing data
y_pred = rf_clf.predict(X_test)

# Classification Report
print(classification_report(y_pred, y_test))
```

```python
shap.initjs()
explainer = shap.TreeExplainer(rf_clf)
shap_values = explainer(X_test)
```

```python
# Class 1Outcome=1)
shap.plots.bar(shap_values[..., 1])
```

```python
# 100th record
shap.plots.bar(shap_values[100][..., 1])
```

```python
# Class 1Outcome=1)
shap.plots.beeswarm(shap_values[..., 1])
```

```python
explanation = explainer(X_test)
shap.plots.scatter(explanation[:, 'density', 1])
```

```python
from lime.lime_tabular import LimeTabularExplainer

# Get the class names
class_names = ['High Quality Wine', 'Low Quality Wine']

# Get the feature names
feature_names = list(X_train.columns)

# Fit the Explainer on the training data set using the LimeTabularExplainer
explainer = LimeTabularExplainer(X_train.values, feature_names =
                                 feature_names,
                                 class_names = class_names,
                                 mode = 'classification')
```

```python
# Perform the explanation on the 8th instance in the test data
explanation = explainer.explain_instance(X_test.iloc[1], rf_clf.predict_proba)

# For PDF export compatibility, use matplotlib to create static visualizations
import matplotlib.pyplot as plt

# Get the feature importance data
exp_list = explanation.as_list()
features = [x[0] for x in exp_list]
scores = [x[1] for x in exp_list]

# Sort by absolute value for better visualization
sorted_indices = [i for i, _ in sorted(enumerate(scores), key=lambda x: abs(x[1]), reverse=True)]
sorted_features = [features[i] for i in sorted_indices]
sorted_scores = [scores[i] for i in sorted_indices]

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
colors = ['green' if s > 0 else 'red' for s in sorted_scores]
plt.barh(range(len(sorted_features)), sorted_scores, color=colors)
plt.yticks(range(len(sorted_features)), sorted_features)
plt.xlabel('Impact on prediction')
plt.title(f'LIME explanation for instance {8}')
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.tight_layout()
plt.show()

# If you need the prediction probability
prediction_proba = rf_clf.predict_proba(X_test.iloc[8].values.reshape(1, -1))
print(f"Prediction probabilities: {prediction_proba}")
```
